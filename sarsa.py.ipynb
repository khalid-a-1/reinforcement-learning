# Reinforcement learning: project 4
## SARSA implementation


```python
import numpy as np
import random

"""making the game board and policy states"""

gamma = 0.1 
rewardSize = 1
gs = 10
states = [[i, j] for i in range(gs) for j in range(gs)]

top_states = [[0, i] for i in range(1, 9)]
bot_states = [[9, i] for i in range(1, 9)]
left_states = [[i, 0] for i in range(1, 9)]
right_states = [[i, 9] for i in range(1, 9)]
start_state = [[0, 0]]
top_right_state = [[0, 9]]
bot_right_state = [[9, 9]]
winning_states = [[0, 9]]
endState = [0, 0]
actionz = {'L': [0, -1], 'R': [0, 1], 'U': [-1, 0], 'D': [1, 0], 'Exit': [0, 0]}
start= ['R', 'U']
general = ['L', 'R', 'U', 'D']
right = ['L', 'U', 'D']
left = ['R', 'U', 'D']
top = ['D', 'L', 'R']
bot = ['U', 'L', 'R']
top_right = ['L', 'D']
bot_right = ['L', 'U']
enda = ['Exit']
numIterations = 1000
l1 = [[0, 0], [2, 3]]
l2 = [[3, 0], [6, 1]]
l3 = [[8, 0], [9, 3]]
l4 = [[7, 2], [3, 8]]
l5 = [[4, 3], [3, 4]] 
l6 = [[0, 2], [1, 4]]
l7 = [[9, 5], [6, 6]]
l8 = [[0, 7], [0, 9]]
l9 = [[9, 7], [9, 9]]

s = [[2, 7], [2, 9]]
s1 = [[1, 1], [1, 6]]
s2 = [[0, 5], [3, 6]]
s3 = [[3, 2], [6, 8]]
s4 = [[5, 7], [5, 9]]
s5 = [[7, 7], [7, 9]]
s6 = [[7, 5], [4, 5]]
s7 = [[5, 2], [7, 4]]
s8 = [[9, 1], [8, 4]]
s9 = [[5, 0], [4, 1]]

"""
This function is inplementing the policy for SARSA for chutes and ladders.
"""

def policy(initialPosition, action, reward_R):   
    
    dicecolors = { "black": 2/3, "black1" :-2/3, "red": 1, "red1": -2/3, "red2": -1/3, 
                  "green": 1, "green1":-1/2, "green2": -1/2, "blue": 1, "blue1": -1}
    key, val = random.choice(list(dicecolors.items())) 
    if dicecolors[key] == 'black' or dicecolors[key] == 'black1':
        reward_R = reward_R + dicecolors[value]
    elif dicecolors[key] == 'red' or dicecolors[key] == 'red1' or dicecolors[key] == 'red2':
        reward_R = reward_R + dicecolors[value]
    elif dicecolors[key] == 'green' or dicecolors[key] == 'green1' or dicecolors[key] == 'green2':
        reward_R = reward_R + dicecolors[value]
    elif dicecolors[key] == 'blue' or dicecolors[key] == 'blue1':
        reward_R = reward_R + dicecolors[value]
        
    if initialPosition == endState:
        return initialPosition, 0
    elif initialPosition == s[0]:
        return s[1], reward_R-1
    elif initialPosition == s1[0]:
        return s1[1], reward_R-1
    elif initialPosition == s2[0]:
        return s2[1], reward_R-1
    elif initialPosition == s3[0]:
        return s3[1], reward_R-1
    elif initialPosition == s4[0]:
        return s4[1], reward_R-1
    elif initialPosition == s5[0]:
        return s5[1], reward_R-1
    elif initialPosition == s6[0]:
        return s6[1], reward_R-1
    elif initialPosition == s7[0]:
        return s7[1], reward_R-1
    elif initialPosition == s7[0]:
        return s7[1], reward_R-1
    elif initialPosition == s8[0]:
        return s8[1], reward_R-1
    elif initialPosition == s9[0]:
        return s9[1], reward_R-1
    elif initialPosition == l1[0]:
        return l1[1], reward_R+1
    elif initialPosition == l2[0]:
        return l2[1], reward_R+1
    elif initialPosition == l3[0]:
        return l3[1], reward_R+1
    elif initialPosition == l4[0]:
        return l4[1], reward_R+1
    elif initialPosition == l5[0]:
        return l5[1], reward_R+1
    elif initialPosition == l6[0]:
        return l6[1], reward_R+1
    elif initialPosition == l7[0]:
        return l7[1], reward_R+1
    elif initialPosition == l8[0]:
        return l8[1], reward_R+1
    elif initialPosition == l9[0]:
        return l9[1], reward_R+1

    finalPosition = initialPosition + actionz[action]
    
    if finalPosition == endState:
        return initialPosition, 0
    elif finalPosition == s[0]:
        return s[1], reward_R-1
    elif finalPosition == s1[0]:
        return s1[1], reward_R-1
    elif finalPosition == s2[0]:
        return s2[1], reward_R-1
    elif finalPosition == s3[0]:
        return s3[1], reward_R-1
    elif finalPosition == s4[0]:
        return s4[1], reward_R-1
    elif finalPosition == s5[0]:
        return s5[1], reward_R-1
    elif finalPosition == s6[0]:
        return s6[1], reward_R-1
    elif finalPosition == s7[0]:
        return s7[1], reward_R-1
    elif finalPosition == s7[0]:
        return s7[1], reward_R-1
    elif finalPosition == s8[0]:
        return s8[1], reward_R-1
    elif finalPosition == s9[0]:
        return s9[1], reward_R-1
    elif finalPosition == l1[0]:
        return l1[1], reward_R+1
    elif finalPosition == l2[0]:
        return l2[1], reward_R+1
    elif finalPosition == l3[0]:
        return l3[1], reward_R+1
    elif finalPosition == l4[0]:
        return l4[1], reward_R+1
    elif finalPosition == l5[0]:
        return l5[1], reward_R+1
    elif finalPosition == l6[0]:
        return l6[1], reward_R+1
    elif finalPosition == l7[0]:
        return l7[1], reward_R+1
    elif finalPosition == l8[0]:
        return l8[1], reward_R+1
    elif finalPosition == l9[0]:
        return l9[1], reward_R+1
    elif finalPosition == endState:
        return endState, rewardSize
    elif -1 in finalPosition or 9 in finalPosition:
        finalPosition = initialPosition
        return finalPosition, 0
    else:
        return finalPosition, rewardSize
    
```

```python
"""
this function calculates the correspongding matrices values for each state and policy updates.
"""
def values(vals, state, reward_R, verbose=False):
    hexs = []
    opt = np.empty((gs, gs), dtype=str)
    for it in range(1000):
        cvp = np.copy(vals)
        hS = []
        for state in states:
            w = 0
            if state in start_state:
                actions = start
            elif state in top_states:
                actions = top
            elif state in bot_states:
                actions = bot
            elif state in left_states:
                actions = left
            elif state in right_states:
                actions = right
            elif state in top_right_state:
                actions = top_right
            elif state in bot_right_state:
                actions = bot_right
            elif state == endState:
                actions = enda
            else:
                actions = general
            q = []
            for action in actions:
                finalPosition, reward = policy(state, action, reward_R)
                expected_return = (1/len(actions))*(reward+(gamma* (vals[finalPosition[0], finalPosition[1]])))
                q.append((expected_return, action))
            m, n = max(q)
            hS.append(np.abs(cvp[state[0], state[1]] - m))
            cvp[state[0], state[1]] = m
            opt[state[0], state[1]] = n
        hexs.append(hS)
        vals = cvp
        if verbose==True:
            print("Iteration {}".format(it))
            print(vals)
            print()
        if max(hS) < 0.001:
            return vals, opt
    
```

```python
value, policy  = values(np.zeros((gs, gs)), state = [0,0], reward_R=195, verbose=False)
print(value)
print(policy)
```

```python
"""
This function shows the average number of dice rolls needed to reach optimal path.
"""
def main():
    sums = 0
    count = 0
    for v in value:
        for x in v:
            count +=1
            sums = sums + x
    print(sums/count)
```
